```python
import matplotlib.pyplot as plt
import numpy as np

# Forecasting Models
models = ['ARIMA (Fourier)', 'ARIMA+Fourier', 'TBATS']

# Error Metrics
metrics = ['MAE', 'RMSE']
values = {
    'MAE': [11.58, 11.58, 9.33],
    'RMSE': [20.50, 20.50, 19.67]
}

x = np.arange(len(metrics))
bar_width = 0.25

plt.figure(figsize=(10, 6))

# Plot bars for each model grouped by metric
for i, model in enumerate(models):
    metric_vals = [values[metric][i] for metric in metrics]
    plt.bar(x + (i - 1) * bar_width, metric_vals, width=bar_width, label=model)

# Add labels
plt.xticks(x, metrics)
plt.ylabel("Error Value")
plt.title("(2024) (m = 12) ARIMA + ARIMA with fourier terms + TBATS")
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Annotate values on top of bars
for i, metric in enumerate(metrics):
    for j, model in enumerate(models):
        val = values[metric][j]
        x_pos = i + (j - 1) * bar_width
        plt.text(x_pos, val + 0.3, f"{val:.2f}", ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = ['ARIMA (Fourier)', 'ARIMA+Fourier', 'TBATS']

# Metric values
mae = [0.7834, 0.7220, 0.7226]
mse = [4.6509, 4.7828, 4.7817]
rmse = [2.1566, 2.1870, 2.1867]
r2 = [-0.0708, -0.1012, -0.1009]

# Setup plot
x = np.arange(len(models))
bar_width = 0.2

plt.figure(figsize=(14, 6))

# Plot bars
plt.bar(x - 1.5*bar_width, mae, width=bar_width, label='MAE', color='skyblue')
plt.bar(x - 0.5*bar_width, mse, width=bar_width, label='MSE', color='gold')
plt.bar(x + 0.5*bar_width, rmse, width=bar_width, label='RMSE', color='salmon')
plt.bar(x + 1.5*bar_width, r2, width=bar_width, label='R² Score', color='lightgreen')

# Labels and title
plt.xticks(x, models)
plt.ylabel("Metric Value")
plt.title("(2023) (m = 12) ARIMA + ARIMA with fourier terms + TBATS")
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Annotate values
for i in range(len(models)):
    plt.text(x[i] - 1.5*bar_width, mae[i] + 0.05, f"{mae[i]:.3f}", ha='center', fontsize=8)
    plt.text(x[i] - 0.5*bar_width, mse[i] + 0.05, f"{mse[i]:.3f}", ha='center', fontsize=8)
    plt.text(x[i] + 0.5*bar_width, rmse[i] + 0.05, f"{rmse[i]:.3f}", ha='center', fontsize=8)
    plt.text(x[i] + 1.5*bar_width, r2[i] + 0.05 if r2[i] > 0 else r2[i] - 0.15, f"{r2[i]:.3f}", ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics
metrics = ['MAE', 'MSE', 'RMSE', 'R²']
models = ['ARIMA (Fourier)', 'ARIMA+Fourier', 'TBATS']

# Values for each metric
values = {
    'MAE': [0.7834, 0.7220, 0.7226],
    'MSE': [4.6509, 4.7828, 4.7817],
    'RMSE': [2.1566, 2.1870, 2.1867],
    'R²': [-0.0708, -0.1012, -0.1009]
}

# Set up the bar chart
x = np.arange(len(metrics))
bar_width = 0.2

plt.figure(figsize=(12, 6))

# Plot each model's performance across metrics
for i, model in enumerate(models):
    model_values = [values[m][i] for m in metrics]
    plt.bar(x + (i - 1) * bar_width, model_values, width=bar_width, label=model)

# Labels and style
plt.xticks(x, metrics)
plt.ylabel("Metric Value")
plt.title("(2023) (m = 12) ARIMA + ARIMA with fourier terms + TBATS")
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Annotate bars
for i, metric in enumerate(metrics):
    for j, model in enumerate(models):
        val = values[metric][j]
        y_pos = val + 0.05 if metric != 'R²' or val > 0 else val - 0.15
        plt.text(i + (j - 1) * bar_width, y_pos, f"{val:.3f}", ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```

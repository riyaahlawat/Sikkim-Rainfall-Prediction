```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Ensure inline plotting if using Jupyter Notebook
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
```

```python
import pandas as pd

# 1978 to 2022
file_path = "..\\datasets\\modified_location_0.xlsx"

# Read the Excel file
data = pd.read_excel(file_path)

# Display the first few rows of the data
print(data.head())
```

```python
data['Day_of_Week'] = data['Date'].dt.dayofweek  # Day of the week (0=Monday, 6=Sunday)
data['Day_of_Year'] = data['Date'].dt.dayofyear  # Day of the year (1-365/366) 
```

```python
print(data.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Season'] = label_encoder.fit_transform(data['Season'])
```

```python
X = data.drop(['Landslide_possibility','Location','hasRainfall','Quarter','Date','Day','Rainfall (in mm)'] , axis = 1)
y = data['hasRainfall']
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 1)
```

```python
print(X.columns)
```

```python
# LOGISTIC REGRESSION
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# SUPPORT VECTOR CLASSIFIER
```

```python
from sklearn.linear_model import SGDClassifier
# changed from SVC() then LinearSVC()
model = SGDClassifier(loss="hinge")  # Hinge loss is used for SVM
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# RANDOM FOREST CLASSIFIER
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for MLP across different splits: (80-20, 60-40, 70-30)
mae_sums = [40.367, 43.6022, 40.0561]
mse_sums = [566.759, 702.9841, 663.2411]
r2_sums = [7.981, 8.0098, 7.9181]

# X-axis labels
splits = ['80-20', '60-40', '70-30']

# Create subplots for each metric
fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

# Plot MAE
axs[0].bar(splits, mae_sums, color='skyblue')
axs[0].set_title('MLP MAE across Splits')
axs[0].set_ylabel('MAE')

# Plot MSE
axs[1].bar(splits, mse_sums, color='lightgreen')
axs[1].set_title('MLP MSE across Splits')
axs[1].set_ylabel('MSE')

# Plot R²
axs[2].bar(splits, r2_sums, color='salmon')
axs[2].set_title('MLP R² across Splits')
axs[2].set_ylabel('R² Score')

plt.tight_layout()
plt.show()
```
